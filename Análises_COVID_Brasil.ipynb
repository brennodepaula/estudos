```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(palette = 'Accent', style = 'whitegrid')
```

```python
#importanto diretamente do github
dados = pd.read_csv("https://raw.githubusercontent.com/wcota/covid19br/master/cases-brazil-states.csv")
```

```python
dados.head()
```

```python
dados.info()
```

```python
# mudando date para datetime
dados['date'] = pd.to_datetime(dados['date'])
```

```python
dados.head()
```

## Analisando o estado de goiás

```python
go = dados.query('state == "GO"')
go.head()
```

```python
#Total de Casos por data
ax = sns.lineplot(data = go, x = 'date', y = 'totalCases')
ax.figure.set_size_inches(12, 6)
ax.set_title('Número de Casos de COVID no Estado de Goiás')
ax.set_ylabel('Total de Casos')
plt.ticklabel_format(style = 'plain', axis ='y')
plt.tick_params(labelsize = 14)
ax.set_xlabel('Data')
```

```python
#Ranqueando os meses por número de novos casos

# dt.strftime -> Return an Index of formatted strings specified by date_format, which supports the same string format as the python standard library.
# %b -> mês abreviado / %y -> Ano

go.groupby(go['date'].dt.strftime('%Y/%m'))['newCases'].sum().sort_values(ascending = False)
```

```python
go.query('vaccinated > 0')
```

```python
#Total de Mortes
ax = sns.lineplot(data = go, x = 'date', y = 'deaths')
ax.figure.set_size_inches(12, 6)
plt.ylim(top = 30000)
```

```python
#Total de Novas Mortes
ax = sns.lineplot(data = go, x = 'date', y = 'newDeaths')
ax.figure.set_size_inches(20, 10)
```

```python
#Novas Mortes Por mês
```

```python
df = pd.DataFrame(go.groupby(go['date'].dt.strftime('%Y/%m'))['newDeaths'].sum())
df
```

```python
#Novas mortes por mes
ax = sns.lineplot(data = df[10:], y ='newDeaths', x = df.index[10:], markers = True)
ax.figure.set_size_inches(20, 12)
ax.set_title('Novas Mortes Por mês desde 2021', fontsize = 22)
ax.set_ylabel('Novas Mortes', fontsize = 18)
```

## Comparando alguns valores

```python
centro_oeste = dados.query('state == "MS" or state == "MT" or state == "DF" or state == "GO"')
```

```python
centro_oeste.groupby('state').sum()
```

```python
#comparação do número de casos entre os estados

ax = sns.barplot(data = centro_oeste.groupby('state').sum(), y = 'totalCases', x = centro_oeste.groupby('state').sum().index)
ax.figure.set_size_inches(10, 6)
plt.ylim(top = 350000000)
plt.tick_params(labelsize = 10)
plt.ticklabel_format()
```

```python
#comparação do número de casos entre os estados durante os meses
ax = sns.lineplot(data = centro_oeste, y = 'totalCases', x = 'date', hue = 'state')
ax.figure.set_size_inches(15, 8)
plt.ylim(top = 1500000)
plt.tick_params(labelsize = 12)
```

```python
#comparação do número total de vacinados entre os estados durante os meses
ax = sns.lineplot(data = centro_oeste, y = 'vaccinated', x = 'date', hue = 'state')
ax.figure.set_size_inches(15, 8)
plt.tick_params(labelsize = 12)
```

```python
#comparação do número total de vacinados/hab entre os estados durante os meses
ax = sns.lineplot(data = centro_oeste, y = 'vaccinated_per_100_inhabitants', x = 'date', hue = 'state')
ax.figure.set_size_inches(15, 8)
plt.tick_params(labelsize = 12)
plt.ylim(top = 100)
```
